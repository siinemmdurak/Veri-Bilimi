## Performans ve İşten Ayrılma Analizi
* İşten ayrılma oranını (Attrition) analiz edin ve görselleştirin.
* Departman bazında işten ayrılma oranlarını karşılaştırın.
* Performans değerlendirmesi ve iş tatmininin işten ayrılma üzerindeki etkisini inceleyin.

```python
isten_ayrılma_oran = calisanlar['Attrition'].value_counts(normalize=True).reset_index()
isten_ayrılma_oran # isten ayrılma oranı
```

```python
isten_ayrılma_oran.plot(kind='bar', color=['lightcoral', 'lightgreen'])
plt.title('İşten Ayrılma Oranı (Attrition)')
plt.xlabel('İşten Ayrılma Durumu')
plt.ylabel('Oran')
plt.xticks(rotation=0) 
```

## --------

```python
isten_ayrılma = calisanlar[calisanlar['Attrition'] == 'Yes']['Department'].value_counts()
isten_ayrılma  #departman bazında isten ayrılanların sayısını bulduk
```

```python
isten_ayrılmama = calisanlar[calisanlar['Attrition'] == 'No']['Department'].value_counts()
isten_ayrılmama #departman bazında isten ayrılmayanların sayısını bulduk
```

```python
toplam_calisan = calisanlar.groupby('Department').size()
toplam_calisan #departman bazında toplam calısan sayılarını bulduk
```

```python
ayrılma_oran = isten_ayrılma / toplam_calisan
ayrılma_oran  # isten ayrılanların sayısını toplam calısan sayısına oranlayarak ayrılanların oranını bulduk
```

```python
ayrılmama_oran = isten_ayrılmama / toplam_calisan
ayrılmama_oran  # isten ayrılmayanların sayısını toplam calısan sayısına oranlayarak ayrılmayanların oranını bulduk
```

```python
oran = pd.DataFrame({
    'Ayrılma Oran': ayrılma_oran,
    'Kalma Oran': ayrılmama_oran
})
oran   #Ayrılma ve Kalma oranlarını karşılaştırdım.
```

## Gelir Analizi
* Aylık gelirlerin departmanlara ve iş rollerine göre dağılımını analiz edin.
* Gelir ile iş tatmini veya performans değerlendirmesi arasında bir ilişki olup olmadığını araştırın.

```python
departman_role_gore = calisanlar.groupby(['Department', 'JobRole'])['MonthlyIncome'].mean().reset_index()
departman_role_gore #aylık gelirlerin ortalamasını alarak departman ve iş rollerine göre grupladık
```

## --------

```python
aralik = np.linspace(calisanlar['MonthlyIncome'].min(), calisanlar['MonthlyIncome'].max(), 6)
#MonthlyIncome sütunundaki en düşük ve en yüksek değerler arasında 6 eşit aralığı oluşturur
calisanlar['Gelir'] = pd.cut(calisanlar['MonthlyIncome'], aralik)
#MonthlyIncome sütunundaki değerleri, daha önce oluşturulan aralik dizisini kullanarak
#gelir aralıklarına böler ve bu aralıkları Gelir adlı yeni bir sütunda saklar.
```

```python
gelir_tatmin = calisanlar.groupby('Gelir')['JobSatisfaction'].mean().reset_index()
gelir_tatmin
```

## Mesafe ve İş Tatmini
* Çalışanların evden işe olan mesafesi ile iş tatmini arasında bir ilişki olup olmadığını analiz edin.
* Bu ilişkiyi görselleştirin ve yorumlayın.

```python
aralik1 = np.linspace(calisanlar['DistanceFromHome'].min(), calisanlar['DistanceFromHome'].max(), 6)
calisanlar['Evden İşe Mesafe'] = pd.cut(calisanlar['DistanceFromHome'], aralik1)
mesafe_tatmin = calisanlar.groupby('Evden İşe Mesafe')['JobSatisfaction'].mean().reset_index()
mesafe_tatmin
```

```python
plt.figure(figsize=(10, 6))
plt.bar(mesafe_tatmin['Evden İşe Mesafe'].astype(str), mesafe_tatmin['JobSatisfaction'], color='skyblue')
plt.title('Evden İşe Mesafe ile İş Memnuniyeti Arasındaki İlişki')
plt.xlabel('Evden İşe Mesafe (km)')
plt.ylabel('Ortalama İş Memnuniyeti')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```
